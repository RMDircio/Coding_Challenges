# My Solution

## My Understanding

I will be given a `True` or `False` as a parameter.
I need to return a corresponding `Yes` or `No` for each parameter

- [x] If statement for when True
- [x] Else statement when False
- [x] Should have something for when a boolean is not given
------ Needed a `while` statement for this to work


```python
def bool_to_word(boolean):
    
    # If boolean is given
    while boolean is True or boolean is False:

        if boolean == True:
            return 'Yes'

        if boolean == False:
            return 'No'

    # if boolean is NOT given        
    return "Give Boolean "
```
