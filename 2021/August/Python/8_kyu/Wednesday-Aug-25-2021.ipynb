# My Solution

## My Understanding
I have two values, first is my starting point and the second is my ending point.<br>I take the starting point and I multiply it by 1, then increase the multiplcation by 1.<br>So it would look like `value`*1, `value`*2, `value`*3...so on.<br>I stop when the multiplied value equals the limit that was given.<br>I also need to include that limit if it is multiple of the `value`

First round
    - see if I can take the start and times it by 2
    - stop when I reach the limit
    - I need an empty list to store the multiples


```python
def find_multiples(integer, limit):
    # empty list
    multiplies = []
    #add the start value to the empty list
    multiplies.append(integer)
    number_check = multiplies[-1]
    multiplyer = 1

    for e in multiplies: 
        print(e)
        # if the integer is not greater than the limit
        if e < limit:
            print(f'1st: last# multiplies =', number_check)
            print(f'2nd:',e)
            
            # multiply by 2
            new_value = integer*multiplyer 
            multiplyer += 1
            print(f'New_Value= ', new_value)
            
            # append new value to list
            multiplies.append(new_value)
            print(f'Appeneded New_value', new_value) # forever loop here
            number_check = multiplies[-1]
            print(f'2nd: last# last# multiplies =', number_check)
            print(f'Limit is = ', limit)
            print(f'----------')
    
    # return the new list
    return multiplies
```
